# Step 8 — Baseline vs monitoring crediting deltas

```python
from pathlib import Path
import os, sys, subprocess, json

# Run this notebook from anywhere inside the repo.
REPO = Path.cwd()
while REPO != REPO.parent and not (REPO / 'python').exists():
    REPO = REPO.parent
assert (REPO / 'python').exists(), 'Could not locate repo root. Open the notebook from inside the repo.'
os.chdir(REPO)
print('Repo root:', REPO)
```

```python
BASELINE_CSV = REPO / 'outputs' / 'tables' / 'field_summary_baseline.csv'
MONITORING_CSV = REPO / 'outputs' / 'tables' / 'field_summary_monitoring.csv'
OUTPUT_CSV = REPO / 'outputs' / 'tables' / 'field_crediting_delta.csv'
FLOOR_NEGATIVE_TO_ZERO = True
```

```python
cmd = [
    sys.executable,
    str(REPO / 'python' / '05_crediting_layer.py'),
    '--baseline', str(BASELINE_CSV),
    '--monitoring', str(MONITORING_CSV),
    '--output', str(OUTPUT_CSV),
]
if FLOOR_NEGATIVE_TO_ZERO:
    cmd.append('--floor_negative_to_zero')
subprocess.run(cmd, check=True)
```
