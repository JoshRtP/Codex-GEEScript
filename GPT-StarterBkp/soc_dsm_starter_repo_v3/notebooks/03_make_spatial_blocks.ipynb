# Step 3 — Create explicit spatial blocks

```python
from pathlib import Path
import os, sys, subprocess, json

# Run this notebook from anywhere inside the repo.
REPO = Path.cwd()
while REPO != REPO.parent and not (REPO / 'python').exists():
    REPO = REPO.parent
assert (REPO / 'python').exists(), 'Could not locate repo root. Open the notebook from inside the repo.'
os.chdir(REPO)
print('Repo root:', REPO)
```

```python
INPUT_CSV = REPO / 'outputs' / 'tables' / 'soc_training_table.csv'
OUTPUT_CSV = REPO / 'outputs' / 'tables' / 'soc_training_table_blocked.csv'
METHOD = 'grid'   # or 'kmeans'
GRID_DEG = 0.25
N_CLUSTERS = 25
```

```python
cmd = [
    sys.executable,
    str(REPO / 'python' / '02a_make_spatial_blocks.py'),
    '--input', str(INPUT_CSV),
    '--output', str(OUTPUT_CSV),
    '--method', METHOD,
]
if METHOD == 'grid':
    cmd += ['--grid_deg', str(GRID_DEG)]
else:
    cmd += ['--n_clusters', str(N_CLUSTERS)]
subprocess.run(cmd, check=True)
```
